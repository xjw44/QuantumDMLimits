```python
import yaml
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget

with open('../curves/2512.08065v2/HEPData-ins3091049-v2-SI_cross_section.yaml') as f:
    raw = yaml.safe_load(f)

mass = np.array([v['value'] for v in raw['independent_variables'][0]['values']])

dep = {d['header']['name']: np.array([v['value'] for v in d['values']])
       for d in raw['dependent_variables']}

print("Mass:", mass)
print("Keys:", list(dep.keys()))
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(mass, dep['limit'], color='black', lw=1.5, label='LZ observed')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(3, 10)
ax.set_ylim(1e-46, 1e-41)
ax.set_xlabel(r'DM Mass [GeV/$c^2$]')
ax.set_ylabel(r'DM-nucleon $\sigma_{SI}$ [cm$^2$]')
ax.legend()
ax.grid(True, which='both', alpha=0.3)
plt.tight_layout()
plt.show()
```
